```python
def show_matrix(matrix):
    print('matrix:')
    
    for row in matrix:
        print(row)

        
def get_row(matrix, i):
    """get row"""
#     print('get row')
    
    len_rows = len(matrix)
    
    if 0 <= i and i < len_rows:
        return matrix[i]
    return 'Error: list index out of range'

def change_rows(matrix, src, dest):
    """change rows"""
#     print('change rows')
    
    if src != dest:
        n = len(matrix[0])

        tempRow = [None] * (n - 1)

        tempRow = matrix[dest]

        matrix[dest] = matrix[src]
        matrix[src] = tempRow
        

def get_max_abs_a_i_0(matrix):
    """get max a_i"""
#     print('get max a_i')
    
    n = len(matrix)
    
    max_abs_a_i = abs(matrix[0][0])
    max_index = 0
    
    for i in range(1, n):
        value = abs(matrix[i][0])
#         print(value)
        
        if value > max_abs_a_i:
            max_index = i
            max_abs_a_i = value

    return max_index

def set_max_coef(matrix, max_index):
    """set max coef"""
#     print('set max coef')
    
    row_first_index = 0
    
    if max_index != row_first_index:
        change_rows(matrix, max_index, row_first_index)


def devide_by_a_j(matrix, i):
    n = len(matrix[i])
    a_i = matrix[i][i]
    
    for j in range(n):
        matrix[i][j] /= a_i


def get_extended_matrix(matrix_a, free_b):    
    a_len = len(matrix_a)
    b_len = len(free_b)
    
    if a_len == b_len:
        for i in range(a_len):
            a[i].append(free_b[i])
    
    else:
        print('Error: len of matrix do no equal')
```

```python
def zeroed_x_i(a, prev_row, k):
    n = len(a[0])
    m = len(a)
    
    for i in range(k+1, m):
        a_l = a[i][k]

        for j in range(n):
            a[i][j] = a[i][j] - prev_row[j] * a_l
    
def zeroed_x(a):
    m = len(a) - 1
    
    for i in range(m):
        prev_row = a[i]
        
        zeroed_x_i(a, prev_row, i)
```
