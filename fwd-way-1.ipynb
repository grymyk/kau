```python
def show_matrix(matrix):
    print('matrix:')
    
    for row in matrix:
        print(row)

a = [
    [1, -2, 2, 3],
    [-5, 12, -14, -35],
    [6, -7, 5, 10]
]

def fwd_way(a, k_len):
    m = len(a)
    n = len(a[0])
    
    for k in range(0, k_len):
        print('\nk:', k)

        for i in range(k+1, m):
            print('i:', i)

            temp = [None] * n

            for j in range(k, n):
                print('j:', j)
                
                print(a[k][k], end=' ')
                print(a[i][j], end=' ')
                print(a[k][j], end=' ')
                print(a[i][k], end=' ')
#                 print()

                a_ij = a[k][k] * a[i][j] - a[k][j] * a[i][k]
                temp[j] = a_ij

                print(':', a_ij)

            for j in range(k, n):
                print('j:', j)
                
                a[i][j] = temp[j]

            print()
            show_matrix(a)

k_len = 2

fwd_way(a, k_len)

show_matrix(a)
```
